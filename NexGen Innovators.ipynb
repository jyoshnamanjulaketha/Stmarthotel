```python
# Import necessary libraries
import pandas as pd

# Load the dataset
data = pd.read_csv('hotel_bookings.csv')

# Display the first few rows of the dataset to understand the structure
data.head()

```

```python
# Drop 'required_car_parking_space' due to high missing values
data.drop(columns=['required_car_parking_space'], inplace=True)
```

```python
# Impute missing values with mean for numeric columns
data['no_of_adults'].fillna(data['no_of_adults'].mean(), inplace=True)
data['no_of_children'].fillna(data['no_of_children'].median(), inplace=True)
data['lead_time'].fillna(data['lead_time'].median(), inplace=True)
```

```python
# Impute missing values with mode for categorical columns
data['type_of_meal_plan'].fillna(data['type_of_meal_plan'].mode()[0], inplace=True)
data['room_type_reserved'].fillna(data['room_type_reserved'].mode()[0], inplace=True)
```

```python
# Create an indicator for missing values in 'avg_price_per_room'
data['avg_price_missing'] = data['avg_price_per_room'].isnull().astype(int)
data['avg_price_per_room'].fillna(data['avg_price_per_room'].mean(), inplace=True)
```

```python
# Check for missing values again
print(data.isnull().sum())
```

```python
# Drop columns that are not essential or have excessive missing values
data.drop(columns=['market_segment_type'], inplace=True)

# Impute numeric columns with median
data['no_of_weekend_nights'].fillna(data['no_of_weekend_nights'].median(), inplace=True)
data['no_of_week_nights'].fillna(data['no_of_week_nights'].median(), inplace=True)
data['arrival_year'].fillna(data['arrival_year'].median(), inplace=True)
data['arrival_month'].fillna(data['arrival_month'].median(), inplace=True)
data['arrival_date'].fillna(data['arrival_date'].median(), inplace=True)
data['repeated_guest'].fillna(data['repeated_guest'].median(), inplace=True)
data['no_of_previous_cancellations'].fillna(data['no_of_previous_cancellations'].median(), inplace=True)
data['no_of_previous_bookings_not_canceled'].fillna(data['no_of_previous_bookings_not_canceled'].median(), inplace=True)
data['no_of_special_requests'].fillna(data['no_of_special_requests'].median(), inplace=True)

# Check for remaining missing values
print(data.isnull().sum())
```

```python
# Check for invalid dates using a try-except approach to flag issues
def check_invalid_date(row):
    try:
        pd.to_datetime(f"{int(row['year'])}-{int(row['month'])}-{int(row['day'])}")
        return False  # No error, so it's valid
    except ValueError:
        return True  # Error means the date is invalid

# Apply the function across your data
invalid_dates = data[['year', 'month', 'day']].apply(check_invalid_date, axis=1)

# Show the rows with invalid dates
print(data[invalid_dates])
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import joblib

# Step 1: Ensure the Booking_Date is a datetime object
data['Booking_Date'] = pd.to_datetime(data['Booking_Date'])

# Step 2: Aggregate bookings by month and booking status
monthly_bookings = data.groupby(data['Booking_Date'].dt.to_period('M'))['booking_status'].value_counts().unstack(fill_value=0)

# Create a mapping for booking status
status_mapping = {0: 'Canceled', 1: 'Not_Canceled'}  # Adjust this mapping based on your actual data

# Rename columns based on the mapping
monthly_bookings.rename(columns=status_mapping, inplace=True)

# Check the updated columns of monthly_bookings
print("Updated monthly bookings columns:")
print(monthly_bookings.columns)

# Define the booking status you expect
expected_status = 'Not_Canceled'  # This should match the updated column name

# Ensure the status exists
if expected_status not in monthly_bookings.columns:
    print(f"'{expected_status}' status not found in columns. Available statuses are: {monthly_bookings.columns.tolist()}")
else:
    # Step 6: Define low booking threshold and trigger marketing campaign
    low_booking_threshold = 1000
    low_booking_periods = monthly_bookings[monthly_bookings[expected_status] < low_booking_threshold].dropna()

    # Trigger marketing campaign function
    def trigger_marketing_campaign(low_booking_periods):
        if not low_booking_periods.empty:
            # Reset the index and format Booking_Date to only show month
            low_booking_periods = low_booking_periods.reset_index()
            low_booking_periods['Booking_Date'] = low_booking_periods['Booking_Date'].dt.strftime('%B')  # Get month name
            
            print("Triggering marketing campaign due to low bookings in the following periods:")
            print(low_booking_periods)
        else:
            print("No campaign needed; bookings are adequate.")
    
    # Call the function
    trigger_marketing_campaign(low_booking_periods)

    # Forecasting future bookings using Prophet
    # Prepare data for Prophet
    forecast_data = monthly_bookings.reset_index()[[expected_status]].copy()
    forecast_data['ds'] = monthly_bookings.index.to_timestamp()  # Create date column directly from index
    forecast_data = forecast_data.rename(columns={expected_status: 'y'})  # Rename to 'y' as required by Prophet
    
    # Initialize and fit the Prophet model
    prophet_model = Prophet()
    prophet_model.add_seasonality(name='monthly', period=30.5, fourier_order=5)  # Add monthly seasonality
    prophet_model.fit(forecast_data)

    # Dump the trained model to a file
    joblib.dump(prophet_model, 'prophet_model.pkl')
    print("Model saved successfully as 'prophet_model.pkl'.")

    # Create future dates for forecasting
    future = prophet_model.make_future_dataframe(periods=12, freq='M')  # Forecast for the next 12 months
    forecast = prophet_model.predict(future)

    # Plot the forecast
    fig = prophet_model.plot(forecast)
    plt.title('Booking Forecast for the Next 12 Months')
    plt.xlabel('Date')
    plt.ylabel('Number of Bookings')
    plt.show()

    # Print the forecasted values
    print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(12))  # Display forecast for the next 12 months

    # Check if a marketing campaign is needed based on forecasted values
    print("Checking for marketing campaign need based on forecasted values...")
    for index, row in forecast.iterrows():
        if row['yhat'] < low_booking_threshold:
            print(f"Marketing campaign required for {row['ds'].strftime('%B %Y')}: forecasted bookings = {row['yhat']:.2f}")
        else:
            print(f"No campaign needed for {row['ds'].strftime('%B %Y')}: forecasted bookings = {row['yhat']:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import joblib  # Import joblib for saving the model

# Assume monthly_bookings is already defined and status_mapping is applied correctly
monthly_bookings.rename(columns=status_mapping, inplace=True)

# We will use only 'Not_Canceled' for forecasting future bookings
monthly_bookings_series = monthly_bookings['Not_Canceled']

# Fit ARIMA model (p, d, q) - For simplicity, we use (1, 1, 1) but you should use GridSearch for tuning
model = ARIMA(monthly_bookings_series, order=(1, 1, 1))
model_fit = model.fit()

# Save the fitted ARIMA model
joblib.dump(model_fit, 'arima_model.pkl')
print("ARIMA model saved as 'arima_model.pkl'.")

# Summary of the model
print(model_fit.summary())

# Forecast for the next 4 months
forecast_steps = 4
forecast = model_fit.forecast(steps=forecast_steps)  # Adjust steps for more periods
print("Next month's booking forecast for 4 months:", forecast.values)

# Create new indices for the forecasted values
last_period_index = monthly_bookings_series.index[-1]  # Get the last period index
next_month_indices = [last_period_index + i for i in range(1, forecast_steps + 1)]  # Generate indices for next 4 months
forecast_series = pd.Series(forecast.values, index=next_month_indices)

# Plot the actual bookings and the forecasted data
plt.figure(figsize=(10, 5))  # Adjust figure size if necessary
monthly_bookings_series.plot(label='Actual Bookings', color='blue')
forecast_series.plot(label='Forecasted Bookings', color='red', marker='o')
plt.axvline(x=next_month_indices[0], color='gray', linestyle='--', label='Forecast Start')  # Vertical line for reference
plt.title('Booking Forecast for Next 4 Months')
plt.xlabel('Date')
plt.ylabel('Bookings')
plt.legend()
plt.show()

# Save forecasted data to a CSV file if needed
forecast_series.to_csv('forecast_data_arima.csv', header=['Forecasted Bookings'])
print("Forecast data saved as 'forecast_data_arima.csv'.")
```
